```python
import os
import json
import random  # Added for random sampling

def format_local_math_category(base_path, split_folder, category_name):
    """
    Parses the raw text files and injects the category name into the dict.
    """
    file_path = os.path.join(base_path, "mathematics_dataset-v1.0", split_folder, f"{category_name}.txt")
    qa_pairs = []
    
    if not os.path.exists(file_path):
        print(f"File not found: {file_path}")
        return qa_pairs

    with open(file_path, 'r', encoding='utf-8') as f:
        lines = [line.strip() for line in f if line.strip()]
        for i in range(0, len(lines), 2):
            if i + 1 < len(lines):
                qa_pairs.append({
                    "category": category_name, # Added category directly here
                    "question": lines[i],
                    "answer": lines[i+1]
                })
                
    return qa_pairs

# --- 1. Load and aggregate all data ---
base_directory = "/home/ines/Reasoning-activations/src/dataset_processing/math_dataset"
categories_to_format = [
    "arithmetic__add_or_sub", "arithmetic__add_or_sub_in_base",
    "arithmetic__add_sub_multiple", "arithmetic__div",
    "arithmetic__mixed", "arithmetic__mul",
    "arithmetic__mul_div_multiple", "arithmetic__nearest_integer_root",
    "arithmetic__simplify_surd"
]
difficulty_level = "train-hard"

all_pairs = [] # Use a flat list instead of a dictionary

for category in categories_to_format:
    print(f"Formatting {category} from local files...")
    all_pairs.extend(format_local_math_category(base_directory, difficulty_level, category))

# --- 2. Sample 20,000 pairs ---
# We use min() just in case the total dataset happens to have fewer than 20,000 lines
sample_size = min(20000, len(all_pairs))
print(f"Sampling {sample_size} pairs from a total of {len(all_pairs)}...")
sampled_pairs = random.sample(all_pairs, sample_size)

# --- 3. Save to a .jsonl file ---
output_file = "/home/ines/Reasoning-activations/reasoning_datasets/DeppMind-Math_Dataset/formatted_math_samples.jsonl"
print(f"Saving data to {output_file}...")

with open(output_file, 'w', encoding='utf-8') as f_out:
    for pair in sampled_pairs:
        # The pair dict already contains 'category', 'question', and 'answer'
        f_out.write(json.dumps(pair, ensure_ascii=False) + "\n")

print("Done!")
```
